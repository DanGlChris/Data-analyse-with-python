## Use personalized index

* Utiliser la methode set_index pour  indexer  notre DataFrame fandango avec la colonne "Film". A preciser
    que nous souhaitons conserver la colonne "FILM". De plus nous souhaitons garder le DataFrame d'origine et en creer un
    nouveau que nous assignerons à la variable fandango_films.
* Afficher l'index du DataFrame fandango_films à l'aide de l'attribut index.


```python
fandango_films = fandango.set_index(fandango["FILM"], inplace=False, drop=False)
```

## Method Apply() on DataFrame column

* Obtenir le dataframe float_df que j'ai realisé en exemple (qui retourne les colonnes decimale float64)
* Utilisez la methode Apply() sur ce dataframe float_df afin de diviser par 2 chaque valeur et assigner le resultat a la variable exo_df
* Affichez les premiere ligne du resultat

```python
import numpy as np
types = fandango_films.dtypes
float_df = fandango_films[types[types.values == "float64"].index]
exo_df = float_df.apply(lambda x : np.divide(x, 2))
```

```python
print(exo_df)
```

## Method Apply() on DataFrame Row

* Utiliser la methode apply pour calculer la moyenne (methode mean() de la librairie Numpy) des valeurs de chaque film  pour 'RT_user_norm' et 'metacritic_user' puis assigner le résultat à la variable rt_mt_means.
* Afficher 5 premieres valeur de rt_mt_means.

```python
rt_mt_means = float_df[["RT_user_norm", "Metacritic_user_nom"]]
rt_mt_means.apply(lambda x : np.mean(x), axis = 1)
```
