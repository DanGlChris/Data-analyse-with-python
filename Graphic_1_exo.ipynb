## Superposition des courbes

* Définir la zone d'affichage à 6 pouces de large et 3 pouce de haut.
* Générer 2 courbes dans un seul graphique en utilisant la colonne 'MONTH' pour l'axe à la place de la colonne 'DATE';
    > * Une courbe avec les valeurs de 1948 de coleur 'red'
    > * Une courbe avec les valeurs de 1949 de coleur 'blue'

* Utiliser plt.show() pour afficher les courbes 

```python
unrate["MONTH"] = unrate["DATE"].dt.month
fig = plt.figure(figsize = (6, 6))
plt.plot(unrate["MONTH"][:12], unrate["UNRATE"][:12], c = "red")
plt.plot(unrate["MONTH"][12:24], unrate["UNRATE"][12:24], c = "blue") # can edit color
plt.show()
```

## Add more Curves

* Définir la zone d'affichage du graphique à 10 pouces de large et 6 pouces de haut 
* Générer les courbes suivante sur le même graphique:

    > * 1948: courbe en rouge 'red'
    > * 1949: courbe en blue 'blue'
    > * 1950: courbe en vert 'green'
    > * 1951: courbe en orange 'orange'
    > * 1952: courbe en noir 'black'
    
* Utiliser plt.show() pour afficher les courbes

```python
col = ["red", "blue", "green", "orange", "black"]

fig = plt.figure(figsize = (10, 6))
for i in range(5):  
    min_ = i*12
    max_ = (i+1)*12
    plt.plot(unrate["MONTH"][min_: max_], unrate["UNRATE"][min_: max_], c = col[i])
    
plt.show()
```

## Add Legend

* Modifier le code précédent pour inclure une légende precisant l'année pour chaque couleur.
    > * с'est-à-dire que la courbe des données de 1948 qui utilise la couleur rouge 'red' aura comme intitulé dans la légende '1948'
    
* Positionner la légende dans le coin haut et à gauche 'upper left'
* Afficher le graphique

```python
col = ["red", "blue", "green", "orange", "black"]
annee = ["1948", "1949", "1950", "1951", "1952"]

fig = plt.figure(figsize = (10, 6))
for i in range(5):  
    min_ = i*12
    max_ = (i+1)*12
    plt.plot(unrate["MONTH"][min_: max_], unrate["UNRATE"][min_: max_], c = col[i], label = annee[i])
    

plt.legend(loc = "upper left")
plt.show()
```

## Final Curve

* Modifier le code précédent afin:
    > * d'ajouter le titre "Evolution du taux de chômage de 1948 à 1952'
    > * d'ajouter l'intitulé pour l'axe x 'Mois(de 1 à 12)'
    > * d'ajouter l'intitulé pour l'axe y 'Taux de chômage (en%)'

```python
col = ["red", "blue", "green", "orange", "black"]
annee = ["1948", "1949", "1950", "1951", "1952"]

fig = plt.figure(figsize = (10, 6))
for i in range(5):  
    min_ = i*12
    max_ = (i+1)*12
    plt.plot(unrate["MONTH"][min_: max_], unrate["UNRATE"][min_: max_], c = col[i], label = annee[i])
    

plt.legend(loc = "upper left")
plt.title("Evolution du taux de chômage de 1948 à 1952")
plt.xlabel("Mois(de 1 à 12)")
plt.ylabel("Taux de chômage (en%)")
plt.show()
```
