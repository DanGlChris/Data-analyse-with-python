## Graphic comparing

```python
import pandas as pd
unrate  = pd.read_csv("unrate.csv")
unrate["DATE"] = pd.to_datetime(unrate["DATE"])

fig  = plt.figure(figsize = (12, 12))
for i in range(5):
    ax = fig.add_subplot(5, 1, i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index: end_index]
    ax.plot(subset["DATE"], subset["UNRATE"])
    
plt.show()
```

## Superposition des courbes

```python
# pandas.series.dt
unrate["MONTH"] = unrate["DATE"].dt.month
```

```python
# plt.plot()
fig = plt.figure(figsize = (6, 6))
plt.plot(unrate["MONTH"][:12], unrate["UNRATE"][:12], c = "green")
plt.plot(unrate["MONTH"][12:24], unrate["UNRATE"][12:24], c = "Red") # can edit color
plt.show()
```

## Add more Curves

```python
col = ["red", "blue", "green", "orange", "black"]

fig = plt.figure(figsize = (10, 6))
for i in range(5):  
    min_ = i*12
    max_ = (i+1)*12
    plt.plot(unrate["MONTH"][min_: max_], unrate["UNRATE"][min_: max_], c = col[i])
    
plt.show()
```

```python
plt.plot(unrate["MONTH"][:12], unrate["UNRATE"][:12], c = "red", label = "1948")
plt.plot(unrate["MONTH"][12:24], unrate["UNRATE"][12:24], c = "blue", label = "1949")

# plt.legend
plt.legend(loc = "upper left")
```
