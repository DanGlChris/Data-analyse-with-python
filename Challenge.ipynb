##  Number of student by Major category

```python
# Method Serie.unique allow you to get all unique catergorie of serie
# it means that you can group all of series
all_age["Major_category"]
```

```python
total = all_age["Total"]
total_is_null = pd.isnull(total)
total_is_null
```

```python
total_null = total[total_is_null]
total_null
```

```python
## find total of student by category of major
def nb_stud_by_cat(df):
    category = df["Major_category"].unique()
    count_dictionary = {}
    
    for c in category:
        major_df = df[df["Major_category"]==c]
        total = major_df["Total"].sum()
        count_dictionary[c] = total
    return count_dictionary
aa_cat_count = nb_stud_by_cat(all_age)
ra_cat_count = nb_stud_by_cat(recent_grad)
```

```python
aa_cat_count
```

```python
ra_cat_count
```

### Using pivot table method

```python
import numpy
def nb_stud_by_cat_2(df):
    value = df.pivot_table(index="Major_category", values="Total", aggfunc=numpy.sum)
    return value
```

```python
aa_cat_count_2 = nb_stud_by_cat_2(all_age)
aa_cat_count_2
```

```python
ra_cat_count_2 = nb_stud_by_cat_2(recent_grad)
ra_cat_count_2
```

## Compare two dataset

```python
major = recent_grad["Major"].unique()
rg_lower_count = 0
for f in major:
    all_age_row = all_age[all_age["Major"]==f]
    recent_grad_row = recent_grad[recent_grad["Major"]==f]
    
    val_2 = recent_grad_row["Unemployment_rate"].values
    val_1 = all_age_row["Unemployment_rate"].values
    if val_2 < val_1: rg_lower_count +=1
        
print(rg_lower_count)
        
    
```
