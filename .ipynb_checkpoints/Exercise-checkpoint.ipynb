## Find missing value

```python
titanic_survival = pd.read_csv("titanic-survival.csv")
titanic_survival.head()
```

```python
age = titanic_survival["age"]
age_is_null = pd.isnull(age)
age_null = age[age_is_null]
print(age_null)
age_null_count = age_null.size
print(age_null_count)
```

```python
good_ages = age[-age_is_null]
mean_age = good_ages.sum() / good_ages.size
print(mean_age)
```

   ## Simpler Average

```python
mean_fare = titanic_survival["fare"].mean()
print(mean_fare)
```

## Calculate price statistics

```python
fare_by_class = {}
passager_classes = [1, 2, 3]
for i in passager_classes:
    pclass = titanic_survival[titanic_survival["pclass"] == i]
    fare = pclass["fare"]
    fare_mean = fare.mean()
    fare_by_class[i] = fare_mean
print(fare_by_class)
```

## Introduction to Pivot table

```python
import numpy
passager_age = titanic_survival.pivot_table(index="pclass", values= "age", aggfunc=numpy.mean)
print(passager_age)
```

```python
passager_survival = titanic_survival.pivot_table(index="pclass", values= "survived")
print(passager_survival)
```

  ## Pivot Table level 2

```python
port_stats = titanic_survival.pivot_table(index="embarked", values=["fare", "survived"], aggfunc=numpy.sum)
print(port_stats)
```

# Apply function  on DataFrame

```python
## Showing missing element in each column(series)
def null_count(column):
    column_null = pd.isnull(column)
    return column[column_null].size
```

```python
titanic_survival.apply(null_count)
```
