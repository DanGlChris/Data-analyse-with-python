## Use personalized index

```python
# DataFrame.set_index(Series (like index), inplace= true, drop = False) drop remove column that become index 
# when inplace is true, dataframe will be replaced else must be stock somewhere
```

```python
fandango_films = fandango.set_index(fandango["FILM"], inplace=False, drop=False)
```

## Select values from a personalized index

```python
# Select with hocks or method loc[]
fandango_films["Mr. Holmes (2015)": "Two Days, One Night (2014)"]
```

```python
fandango_films.loc["Mr. Holmes (2015)": "Two Days, One Night (2014)"]
```

```python
# specific film
fandango_films.loc["Two Days, One Night (2014)"]
```

```python
# Selecting from film liste
movies = ['Kumiko, The Treasure Hunter (2015)', 'Danny Collins (2015)', 'Ant-Man (2015)']
fandango_films.loc[movies]
```

## Method Apply() on DataFrame column

```python
# Apply() allow to apply function in Series object of DataFrame, in their columns
### Ecart type, is mathematic function that calcul dispersion of datas
## by defaut, Apply() function is for columns, but if we change parameter # axis = 1 or 0), it can be apply in row
```

```python
# dataframe.apply( lambda x : x**2)
```

```python
import numpy as np
```

```python
# we return types of datas like Series
# we want to apply # Ecartype method, but that methode must be applied just with float64 type
```

```python
types = fandango_films.dtypes
types
```

```python
# We filter decimal number that type is float64 and we will return the name of those columns
float_columns = types[types.values == 'float64'].index # it's a list of index from type
float_columns
```

```python
# using hocks for filter columns of condition float_columns
```

```python
float_df = fandango_films[float_columns]
float_df
```

```python
deviations = float_df.apply(lambda x: np.std(x))
```

```python
deviations
```

## Method Apply() on DataFrame Row

```python
# apply(function, axis=1)
rt_mt_user = float_df[["RT_user_norm", "Metacritic_user_nom"]]
```

```python
rt_mt_user
```

```python
rt_mt_user.apply(lambda x : np.std(x), axis=1)
```
