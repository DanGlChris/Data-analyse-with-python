# treat missing values

## Introdaction to DataSet

```python
import pandas as pd
```

```python
titanic_surival = pd.read_csv("titanic-survival.csv")
titanic_surival.head()
```

## Find Missing value

```python
# Pandas.isnull()
```

```python
sex = titanic_surival["sex"]
sex_is_null = pd.isnull(sex)
print(sex_is_null)
```

```python
sex_null = sex[sex_is_null]
print(sex_null)
```

## Problem with missing value

```python
mean_age = sum(titanic_surival["age"]) / len(titanic_surival["age"])
print(mean_age)
```

## Simpler averaging

```python
mean_age = titanic_surival["age"].mean()
print(mean_age)
```

## Introduction to Pivot table

```python
#DataFrame.pivot_tabel()
import numpy
passager_class_fares = titanic_surival.pivot_table(index="pclass", values="fare", aggfunc=numpy.mean)
```

```python
print(passager_class_fares)
```

## Delete missing value

```python
#DataFrame.dropna(axis = 0) axis 0 column, 1 line and subset for specfic case
```

```python
drop_na_rows = titanic_surival.dropna(axis = 0)
print(drop_na_rows)
```

```python
titanic_surival.dropna(axis=0, subset=["name"])
```

# Apply function  on DataFrame

```python
# Dataframe.apply
```

```python
# Fuction that return 100th element
```

```python
def row_100(column):
    # extract 100th element from column
    item = column.iloc[99]
    return item
```

```python
# return 100th element of each column
row_100_var = titanic_surival.apply(row_100)
row_100_var
```

## Apply fuction on lines

```python
## DataFrame.appy(fuction, axis = 1) 1 for apply it on line
```

```python
def is_minor(row):
    if(row["age"]<18):
        return True
    else: return False

minors = titanic_surival.apply(is_minor, axis=1)
#titanic_surival[minors]
minors
```

```python
def which_class(row):
    pclass = row["pclass"]
    if(pd.isnull(pclass)): return "Unknown"
    elif(pclass == 1): return "First Class"
    elif(pclass == 2): return "Second Class"
    else: return "Third Class"
which_class = titanic_surival.apply(which_class, axis=1)
which_class
```
